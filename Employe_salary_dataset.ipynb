```python
import pandas as pd  
df=pd.read_csv(r"C:\Users\ELCOT\Downloads\sample_data.csv.csv")
```

```python
 # Rename columns 
df=df.rename(columns={"AGE":"Age",
                  "Years Exp":"Year Experience"})
df
 
```

```python
# Remove duplicates 
df = df.drop_duplicates()

len(df)
```

```python
# Handle null values (drop/fill) 
df.isnull().sum()
df["Remarks"]=df["Remarks"].fillna("no Remarks")
df
```

```python
# Change data types 


print(df["Age"].to_string())
```

```python
df["Age"]=(df["Age"]
           .str.replace("150","24")
           .str.replace("-5", "25")
           .str.replace("twenty six","26"))
df["Age"]=df["Age"].fillna(25)
print(df["Age"].to_string())

```

```python
# Standardize skills
df["Skill Set"] = (
    df["Skill Set"]
        .str.lower()
        .str.upper()
        .str.strip()          
        .str.title()          
        .str.replace("-", " "))
df
```

```python
# Clean salary values 
print(df["Salary (INR)"].to_string())  
```

```python
# clean salary value

import numpy as np
df["Salary (INR)"] = (
    df["Salary (INR)"]
        .str.replace(",","")
        .str.replace("650000 INR","650000")
        .str.replace("$500000","500000")
        .str.replace("4.8L","4800000")
        .str.replace("Nine Lakhs","900000"))


df["Salary (INR)"]=df["Salary (INR)"].fillna(50000)          

print(df["Salary (INR)"].to_string())
```
